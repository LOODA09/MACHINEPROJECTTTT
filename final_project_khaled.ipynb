```python
# For every complementary booking ID, how many non-canceled reservations happened until the next complementary booking?
```

```python
complementary_df = (df[df["market segment type"] == "Complementary"].sort_values("date of reservation").reset_index(drop=True))
counts = []
for i in range(len(complementary_df) - 1):
    comp_id = complementary_df.loc[i, "Booking_ID"]
    start_date = complementary_df.loc[i, "date of reservation"]
    end_date = complementary_df.loc[i + 1, "date of reservation"]
    
    total_reservations = df[
        (df["date of reservation"] > start_date)
        & (df["date of reservation"] < end_date)
    ]["Booking_ID"].count()  
    
    counts.append({
        "Complementary Booking ID": comp_id,
        "Total Reservations Until Next": total_reservations
    })
visual_df = pd.DataFrame(counts)

px.bar(
    visual_df,
    x="Complementary Booking ID",
    y="Total Reservations Until Next",
    title="All Reservations Between Complementary Bookings",
    color="Total Reservations Until Next",
    color_continuous_scale="tempo"
)

```

# --- REFACTORED MODELING SECTION BY ANTIGRAVITY ---


```python
import pandas as pd
import numpy as np
import pickle
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.cluster import KMeans
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
# Import TF/Keras - Ensure you are in a Python 3.8-3.11 environment
try:
    import tensorflow as tf
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense, LSTM, SimpleRNN
    HAS_TF = True
except ImportError:
    HAS_TF = False
    print("TensorFlow not found. Please install it to run ANN/RNN/LSTM.")

# Load Dataset
df_khaled = pd.read_csv('KHALED .csv')
df_khaled.drop(['Booking_ID', 'date of reservation'], axis=1, inplace=True, errors='ignore')

# Preprocessing
le = LabelEncoder()
for col in df_khaled.select_dtypes(include=['object']).columns:
    if col != 'booking status':
        df_khaled[col] = le.fit_transform(df_khaled[col])
df_khaled['booking status'] = df_khaled['booking status'].apply(lambda x: 1 if x == 'Not_Canceled' else 0)

X = df_khaled.drop('booking status', axis=1)
y = df_khaled['booking status']

scaler_khaled = StandardScaler()
X_scaled = scaler_khaled.fit_transform(X)
pickle.dump(scaler_khaled, open('scaler_khaled.pkl', 'wb'))

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

```

## K-Means Segmentation


```python
# K-Means Segmentation
kmeans_khaled = KMeans(n_clusters=3, random_state=42)
df_khaled['segment'] = kmeans_khaled.fit_predict(X_scaled)
print("K-Means segmentation completed.")
```

## Random Forest (Khaled)


```python
# Random Forest
khaled_rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
khaled_rf_cv_scores = cross_val_score(khaled_rf_model, X_scaled, y, cv=kf)
khaled_rf_model.fit(X_train, y_train)
pickle.dump(khaled_rf_model, open('rf_model_khaled.pkl', 'wb'))
print(f"Random Forest CV Mean Accuracy: {khaled_rf_cv_scores.mean():.4f}")
```

## KNN


```python
# KNN
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_cv_scores = cross_val_score(knn_model, X_scaled, y, cv=kf)
knn_model.fit(X_train, y_train)
pickle.dump(knn_model, open('knn_model_khaled.pkl', 'wb'))
print(f"KNN CV Mean Accuracy: {knn_cv_scores.mean():.4f}")
```

## Logistic Regression


```python
# Logistic Regression
lr_model = LogisticRegression(max_iter=1000)
lr_cv_scores = cross_val_score(lr_model, X_scaled, y, cv=kf)
lr_model.fit(X_train, y_train)
pickle.dump(lr_model, open('lr_model_khaled.pkl', 'wb'))
print(f"Logistic Regression CV Mean Accuracy: {lr_cv_scores.mean():.4f}")
```

## XGBoost


```python
# XGBoost
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_cv_scores = cross_val_score(xgb_model, X_scaled, y, cv=kf)
xgb_model.fit(X_train, y_train)
pickle.dump(xgb_model, open('xgb_model_khaled.pkl', 'wb'))
print(f"XGBoost CV Mean Accuracy: {xgb_cv_scores.mean():.4f}")
```

## ANN (Khaled)


```python
# ANN
def create_ann():
    model = Sequential([
        Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

if HAS_TF:
    khaled_ann_model = create_ann()
    khaled_ann_cv_results = []
    for train_idx, val_idx in kf.split(X_scaled):
        m = create_ann()
        m.fit(X_scaled[train_idx], y.iloc[train_idx], epochs=5, batch_size=32, verbose=0)
        khaled_ann_cv_results.append(m.evaluate(X_scaled[val_idx], y.iloc[val_idx], verbose=0)[1])
    
    khaled_ann_model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)
    khaled_ann_model.save('ann_model_khaled.h5')
    print(f"ANN CV Mean Accuracy: {np.mean(khaled_ann_cv_results):.4f}")
else:
    print("Skipping ANN training (TensorFlow missing).")
```

## RNN (Khaled)


```python
# RNN
if HAS_TF:
    X_rnn = X_scaled.reshape((X_scaled.shape[0], 1, X_scaled.shape[1]))
    X_train_rnn = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))

    def create_rnn():
        model = Sequential([
            SimpleRNN(32, input_shape=(1, X_train.shape[1])),
            Dense(1, activation='sigmoid')
        ])
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        return model

    khaled_rnn_model = create_rnn()
    khaled_rnn_cv_results = []
    for train_idx, val_idx in kf.split(X_rnn):
        m = create_rnn()
        m.fit(X_rnn[train_idx], y.iloc[train_idx], epochs=5, verbose=0)
        khaled_rnn_cv_results.append(m.evaluate(X_rnn[val_idx], y.iloc[val_idx], verbose=0)[1])

    khaled_rnn_model.fit(X_train_rnn, y_train, epochs=10, verbose=0)
    khaled_rnn_model.save('rnn_model_khaled.h5')
    print(f"RNN CV Mean Accuracy: {np.mean(khaled_rnn_cv_results):.4f}")
else:
    print("Skipping RNN training (TensorFlow missing).")
```

## LSTM


```python
# LSTM
if HAS_TF:
    def create_lstm():
        model = Sequential([
            LSTM(32, input_shape=(1, X_train.shape[1])),
            Dense(1, activation='sigmoid')
        ])
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        return model

    lstm_model = create_lstm()
    lstm_cv_results = []
    for train_idx, val_idx in kf.split(X_rnn):
        m = create_lstm()
        m.fit(X_rnn[train_idx], y.iloc[train_idx], epochs=5, verbose=0)
        lstm_cv_results.append(m.evaluate(X_rnn[val_idx], y.iloc[val_idx], verbose=0)[1])

    lstm_model.fit(X_train_rnn, y_train, epochs=10, verbose=0)
    lstm_model.save('lstm_model_khaled.h5')
    print(f"LSTM CV Mean Accuracy: {np.mean(lstm_cv_results):.4f}")
else:
    print("Skipping LSTM training (TensorFlow missing).")
```
